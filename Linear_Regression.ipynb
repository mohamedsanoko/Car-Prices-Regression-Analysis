## Importing the relevant libraries

```python
import numpy as np
import pandas as pd 
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set()
```

## Loading the raw data

```python
raw_data = pd.read_csv('car+sales.csv')
raw_data.head()
```

## Preprocessing

### Exploring the descriptive statistics of the variables

```python
raw_data.describe(include='all')
```

### Determing the variables of interest

```python
data = raw_data.drop(['Model'], axis=1)
data.describe(include='all')
```

### Dealing with oultiers

```python
q = data_no_mv['Price'].quantile(0.99)
data_1 = data_no_mv[data_no_mv['Price'] < q]
sns.histplot(data_1['Price'])
```

```python
q = data_1['Mileage'].quantile(0.99)
data_2 = data_1[data_1['Mileage'] < q]
sns.histplot(data_2['Mileage'])
```

```python
data_3 = data_2[data_2['EngineV'] < 6.5]
sns.histplot(data_3['EngineV'])
```

After doing some research, I found that the engine volume varies from 1 to 6.5.

```python
q = data_3['Year'].quantile(0.01)
data_4 = data_3[data_3['Year'] > q]
sns.histplot(data_4['Year'])
```

```python
data_cleaned = data_4.reset_index(drop=True)
```

```python
data_cleaned.describe(include='all')
```

```python
data_cleaned = data_cleaned.drop(['Price'], axis=1)
```

### Multicollinearity

```python
data_cleaned.columns.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage','Year','EngineV']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif['features'] = variables.columns
```

```python
vif
```

```python
data_no_multicollinearity = data_cleaned.drop(['Year'], axis=1)
data_no_multicollinearity
```

## Create dummy variables for categorical data

```python
data_with_dummies = pd.get_dummies(data_no_multicollinearity, drop_first=True, dtype=int)
data_with_dummies.head()
```

### Rearrange a bit

```python
data_with_dummies.columns.values
```

```python
cols = ['log_price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed.head()
```

## Linear regression model

### Declare the inputs and the targets

```python
targets = data_preprocessed['log_price']
inputs = data_preprocessed.drop(['log_price'], axis=1)
```

### Scale the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(inputs)
```

```python
inputs_scaled = scaler.transform(inputs)
inputs_scaled.shape
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(inputs_scaled, targets, test_size=0.2, random_state=365)
```

### Create the regression

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_hat = reg.predict(x_train)
```

```python
plt.scatter(y_train,y_hat)
plt.xlabel('Targets - y_train',fontsize=18)
plt.ylabel('Predicted - y_hat',fontsize=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
sns.histplot(y_train - y_hat)
plt.title("Residuals PDF", size=18)
plt.show()
```

```python
reg.score(x_train,y_train)
```

### Finding the weight and bias

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg_summary = pd.DataFrame(inputs.columns.values, columns=['Features'])
reg_summary['Weights'] = reg.coef_
reg_summary
```

## Testing

```python
y_hat_test = reg.predict(x_test)
```

```python
plt.scatter(y_test,y_hat_test, alpha=0.2)
plt.xlabel('Targets - y_test',fontsize=18)
plt.ylabel('Predicted - y_hat_test',fontsize=18)
plt.plot([6,13], [6,13], color='orange', lw=4, label="45-degree line")
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Predictions'])
df_pf.head()
```

```python
y_test = y_test.reset_index(drop=True)
y_test.head()
```

```python
df_pf = df_pf.drop(['Target'], axis=1)
```

```python
df_pf['Residual'] = df_pf['Targets'] - df_pf['Predictions']
df_pf
```

```python
df_pf['Difference%'] = np.abs(df_pf['Residual']/df_pf['Targets']*100)
```
